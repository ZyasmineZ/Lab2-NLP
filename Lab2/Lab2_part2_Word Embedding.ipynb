```python
import pandas as pd

# Load data from the JSON file
data = pd.read_json("cleaned_data.json")

# Display the loaded data
data
```

```python
# one-hot encoding
one_hot_encoded_data = pd.get_dummies(data)

# Display the one-hot encoded data
one_hot_encoded_data
```

```python
# Bag of words
from sklearn.feature_extraction.text import CountVectorizer

# Concatenate all words in each row to form phrases
phrases = [" ".join(row.dropna()) for _, row in data.iterrows()]

# Initialize CountVectorizer
count_vectorizer = CountVectorizer()

# Apply bag of words transformation
bag_of_words = count_vectorizer.fit_transform(phrases)

# Convert the transformed data into a DataFrame for easier visualization
bag_of_words_df = pd.DataFrame(bag_of_words.toarray(), columns=count_vectorizer.get_feature_names_out())

# Display the bag of words DataFrame
bag_of_words_df
```

```python
# TF/IDF
from sklearn.feature_extraction.text import TfidfVectorizer

# Concatenate all words in each row to form phrases
phrases = [" ".join(row.dropna()) for _, row in data.iterrows()]

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Apply TF-IDF transformation to the data
tfidf_matrix = tfidf_vectorizer.fit_transform(phrases)

# Convert the TF-IDF matrix into a DataFrame for easier visualization
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Display the TF-IDF DataFrame
tfidf_df
```

```python
#CBOW
from gensim.models import Word2Vec
import nltk

# Concatenate all words in each row to form phrases
phrases = [" ".join(row.dropna()) for _, row in data.iterrows()]

# Tokenize phrases into words
tokenized_phrases = [nltk.word_tokenize(phrase) for phrase in phrases]

# Train CBOW model
cbow_model = Word2Vec(sentences=tokenized_phrases, vector_size=100, window=5, sg=0, min_count=1, workers=4)

# Get word vectors
word_vectors = cbow_model.wv

# Example: Get word vector for a specific word
word_vector = word_vectors["الشمس"]

# Print the word vector
word_vector
```

```python
# GloVe Approach
from gensim.models import KeyedVectors

file_path = 'CBOW_500_5_400/w2v_CBOW_500_5_400_10.model'

# Concatenate all words in each row to form phrases
phrases = [" ".join(row.dropna()) for _, row in data.iterrows()]

# Tokenize phrases into words
tokenized_phrases = [phrase.split() for phrase in phrases]

word_embed = KeyedVectors.load_word2vec_format(file_path, encoding='iso-8859-6')

# Example: Get word vector for a specific word
word_vector = word_embed["يوم"]

# Print the word vector
word_vector
```

```python
# FastText Approach
from gensim.models import FastText

# Concatenate all words in each row to form phrases
phrases = [" ".join(row.dropna()) for _, row in data.iterrows()]

# Tokenize phrases into words
tokenized_phrases = [phrase.split() for phrase in phrases]

# Train FastText model
fasttext_model = FastText(sentences=tokenized_phrases, vector_size=100, window=5, min_count=1, workers=4, sg=1)

# Example: Get word vector for a specific word
word_vector = fasttext_model.wv["الشمس"]

# Print the word vector
word_vector
```

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import numpy as np

# Reduce dimensionality to a common size (e.g., 100) for all word vectors
common_vector_size = 100
one_hot_encoded_vectors = one_hot_encoded_data.values[:, :common_vector_size]
bag_of_words_vectors = bag_of_words_df.values[:, :common_vector_size]
tfidf_vectors = tfidf_df.values[:, :common_vector_size]
cbow_vectors = cbow_model.wv.vectors[:, :common_vector_size]
fasttext_vectors = fasttext_model.wv.vectors[:, :common_vector_size]

# Concatenate the reduced-size word vectors
all_word_vectors = np.vstack([
    one_hot_encoded_vectors,
    bag_of_words_vectors,
    tfidf_vectors,
    cbow_vectors,
    fasttext_vectors
])

# Apply t-SNE to reduce dimensionality
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(all_word_vectors)

# Plot the embeddings
plt.figure(figsize=(10, 8))
plt.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1], alpha=0.5)
plt.title("t-SNE Visualization of Word Embeddings from Different Models")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```

<h1>Evaluation of Word Representation Approaches</h1>
<p>Evaluating the different approaches for word representation, such as one-hot encoding, Bag of Words (BoW),
    TF-IDF, CBOW, and FastText, involves considering various factors such as their effectiveness in capturing
    semantic meaning, computational efficiency, and applicability to different tasks. Let's discuss each approach
    and provide a general conclusion:</p>

<h2>One-Hot Encoding:</h2>
<p><strong>Pros:</strong> Simple and straightforward encoding method where each word is represented as a
    binary vector.</p>
<p><strong>Cons:</strong> High-dimensional sparse vectors, incapable of capturing semantic similarity
    between words.</p>
<p><strong>Conclusion:</strong> Suitable for tasks where only word presence or absence is relevant, such as
    text classification.</p>

<h2>Bag of Words (BoW):</h2>
<p><strong>Pros:</strong> Represents each document as a frequency vector of word occurrences, capturing
    some context information.</p>
<p><strong>Cons:</strong> Loss of word order and context, inability to capture semantic meaning.</p>
<p><strong>Conclusion:</strong> Useful for text classification and sentiment analysis but less effective
    for tasks requiring semantic understanding.</p>

<h2>TF-IDF:</h2>
<p><strong>Pros:</strong> Improves upon BoW by weighting word frequencies based on their importance in the
    corpus.</p>
<p><strong>Cons:</strong> Still lacks semantic understanding and context.</p>
<p><strong>Conclusion:</strong> Provides better discrimination between common and rare words, enhancing
    performance in tasks like information retrieval and document classification.</p>

<h2>Continuous Bag of Words (CBOW):</h2>
<p><strong>Pros:</strong> Learns distributed representations of words in a continuous vector space,
    capturing semantic similarity.</p>
<p><strong>Cons:</strong> Requires a large amount of training data and may struggle with rare words.</p>
<p><strong>Conclusion:</strong> Effective for capturing semantic relationships between words, suitable for
    tasks like word analogy and semantic similarity.</p>

<h2>GloVe:</h2> <p><strong>Pros:</strong> Utilizes global word co-occurrence statistics to learn vector representations,      capturing semantic relationships between words.</p   <p><strong>Cons:</strong> Relatively slower training process compared to other methods.</    <p><strong>Conclusion:</strong> Effective for capturing semantic similarity and analogy, suitable for vari         NLP tasks including word analogy, word similarity, and sentiment analysis.</p>

<h2>FastText:</h2>
<p><strong>Pros:</strong> Similar to CBOW but also considers subword information, enabling better handling
    of rare and out-of-vocabulary words.</p>
<p><strong>Cons:</strong> Computationally more expensive than CBOW.</p>
<p><strong>Conclusion:</strong> Effective for capturing semantic meaning, especially in morphologically
    rich languages and tasks with a large vocabulary.</p>


<p>In conclusion, the choice of word representation method depends on the specific task requirements, data
    characteristics, and computational resources available. While simpler methods like one-hot encoding and BoW are
    suitable for basic tasks, more advanced techniques like CBOW and FastText offer better semantic understanding
    and are preferable for tasks demanding nuanced language understanding. Additionally, TF-IDF strikes a balance
    between simplicity and effectiveness, making it useful in various text processing tasks.</p>
